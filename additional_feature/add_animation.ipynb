# Animation

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Loading forest fire simulation
with np.load("add_forest_simulation.npz") as n:
    grid = n["state"]
    
# Setting up the initial figure and axes
fig, ax = plt.subplots(constrained_layout=True)
ax.axis("off")

# Set colours for states 0 (empty), 1 (tree) and 2 (fire) respectively
cmap = ListedColormap(["#16060C", "#42692f", "#FF7621"])

# Plot the initial grid
grid_plot_add = ax.imshow(
    grid[0],
    vmin=grid.min(),
    vmax=grid.max(),
    animated=True,
    cmap=cmap
)

# Animating the state of each grid (number of grids dependent on number of steps)
def animate(i):
    grid_plot_add.set_array(grid[i])
    return [grid_plot_add]

# Creating animation object, interval set to 100ms 
anim_add = FuncAnimation(fig, animate, frames=len(grid), interval=100)

# Displaying animation in Jupyter Notebook
HTML(anim_add.to_jshtml())
```
